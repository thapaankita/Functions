###4.  What is the purpose of the return statement in a Python function?


 - The return statement hands a value back to whatever called the function and then stops the function’s execution right there. Without it, the function just runs to the end and implicitly returns None. So you use return when you want the function to produce something—whether that’s a number, a string, an object, or even multiple values in a tuple—that the rest of your code can capture and use.

 -**Example -**

```python
def miles_to_km(miles):
    km = miles * 1.60934
    return km        # send the result back

distance = miles_to_km(5)   # captures 8.0467
print(distance)             # prints 8.0467
```

###5. What are iterators in Python and how do they differ from iterables?
**Iterable:** Any object you can loop over with for … in … — like lists, strings, dictionaries, files, or generators. It must have a method called __iter__() that returns an iterator.

**Iterator:** The object that actually provides the items one at a time. It has two special methods:
  
- __iter__() — returns itself (which means an iterator is also iterable)

- __next__() — returns the next item, or raises StopIteration when there are no more.

```python
# Iterable: a list
my_list = [1, 2, 3]

# Get an iterator from the iterable
my_iter = iter(my_list)

print(my_list)   # This is the iterable itself
print(my_iter)   # This is the iterator object

# Using the iterator to get items one-by-one
print(next(my_iter))  # 1
print(next(my_iter))  # 2
print(next(my_iter))  # 3

# next(my_iter) now would raise StopIteration because items are finished
```

###6.  Explain the concept of generators in Python and how they are defined.
**Generators** in Python are special functions that produce a sequence of values one at a time using the yield keyword. Unlike regular functions that return a single value and exit, generators pause their state after yielding a value and resume when asked for the next one. This makes them really useful for working with large data without using lots of memory.

**Example:**

```python
def countdown(n):
    while n > 0:
        yield n      # pause here and send back current n
        n -= 1      # resume here on next call

for number in countdown(5):
    print(number)
```

###7. What are the advantages of using generators over regular functions?
**Advantages of generators over regular functions:**

- They use less memory because they generate values one at a time instead of creating a full list.

- They compute values only when needed (lazy evaluation), which can be faster.

- They make code simpler when producing sequences.

**Example -**

```python
def count_up(n):
    for i in range(n):
        yield i

for num in count_up(5):
    print(num)
```

###8. What is a lambda function in Python and when is it typically used?
**Lambda functions** are small, anonymous functions defined with the lambda keyword. They can take any number of arguments but contain just one expression — the result of that expression is automatically returned.

They’re usually used for short, simple functions that you want to write quickly without formally defining a full function with def.

###9. Explain the purpose and usage of the `map()` function in Python.
The map() function applies a given function to each item in an iterable (like a list) and returns an iterator with the results. It’s useful when you want to transform or process all items in a collection without writing an explicit loop.

 - **Example-**

```python
numbers = [1, 2, 3, 4]
squares = map(lambda x: x * x, numbers)

print(list(squares))  # Output: [1, 4, 9, 16]
```

###1.  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.


```python
def sum_even(nums):
    """Add up the even numbers in a list."""
    return sum(n for n in nums if n % 2 == 0)
```

###4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    """
    Check if n is prime.
    Only works for 1 <= n <= 200.
    Returns True if prime, False otherwise.
    """
    if not (1 <= n <= 200):
        raise ValueError("Please give me a number between 1 and 200.")
    if n < 2:
        return False
    # test divisors up to sqrt(n)
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
```

###5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class Fibonacci:
    """An iterator that yields the Fibonacci sequence up to `n_terms` numbers."""

    def __init__(self, n_terms):
        if n_terms < 0:
            raise ValueError("Number of terms must be non-negative")
        self.n_terms = n_terms
        self._count = 0        # how many numbers we’ve yielded so far
        self._a, self._b = 0, 1  # starting pair (F0, F1)

    def __iter__(self):
        # An iterator should return itself
        return self

    def __next__(self):
        if self._count >= self.n_terms:
            # Standard way to signal the iterator is exhausted
            raise StopIteration

        # On the first call, _a is 0; then we walk forward
        current = self._a
        self._a, self._b = self._b, self._a + self._b  # advance the pair
        self._count += 1
        return current
```

###6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(max_exponent):
    """Yield 2**k for k = 0 … max_exponent (inclusive)."""
    if max_exponent < 0:
        raise ValueError("Exponent must be non-negative")

    value = 1         # 2**0
    for _ in range(max_exponent + 1):
        yield value
        value *= 2     # next power (multiply by 2 each step)
```

###7.  Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_lines(path):
    """Yield each line from a text file, one at a time."""
    with open(path, encoding="utf-8") as f:
        for line in f:
            yield line.rstrip("\n")  # drop trailing newline
```
